```python
############ ATIVIDADE 1 ############

############ 1 - retirar espaço ############

nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM"

print(nome.strip())
print(email.strip())

############ 2 - Converter para minusculo ############
print(nome.lower())
print(email.lower())

############ 3 substituir o email ############
email = email.replace("EMPRESA.COM","gmail.com")
print(email)
############ 4 Qtd de caracteres ############
quantidade = len(nome)
print(quantidade)
############ 5 - ultimo/primeira
ultima_letra = nome[-1]
print(ultima_letra)
primeira_letra = nome[0]
print(primeira_letra)
```

```python
###################### ATIVIDADE 2 ######################
produto = input("Digite seu produto: ")
categoria = input("Digite sua categoria: ")
codigo = input("Digite seu codigo: ")
produto = produto.strip()
categoria = categoria.strip()
codigo = codigo.strip()

#2 - Categoria minúsculo
categoria = categoria.lower()
#3 - Maiúsculo e Substituir
codigo = codigo.upper()
codigo = codigo.replace("-", " ")
#4 - Quantidade de caracteres
quantidade = len(produto)
#5 - Primeira letra e 4 preimeiros caracteres do codigo
primeira_letra = produto[0]
carac_cod = codigo[:4]
print("Produto: ",produto)
print("Categoria: ",categoria)
print("Código: ",codigo)
print("Quantidade de caracteres: ",quantidade)
print("Primeira letra: ",primeira_letra)
print("4 primeiros caracteres do código: ",carac_cod)
```
